## 1. 시계열 도표

> 먼저 시계열 도표를 그려보자. 연습을 위해 'global_internet_users.csv'를 불러와 내용을 살펴보자.

````python
import pandas as pd

df = pd.read_csv('global_internet_users.csv')
df

#어떤 나라들이 있는지 살펴보자.
df['Entity'].unique()
````

> **연습문제 1-1**. 한국(`South Korea`), 중국(`China`), 인도(`India`) 레코드를 필터링하여 출력하시오.

````python
cond1 = df['Entity']=='South Korea'
cond2 = df['Entity']=='China'
cond3 = df['Entity']=='India'
df.loc[cond1 | cond2 | cond3]
````

> 그런데 좀 더 편리하고 유용한 매서드가 있는데, 그건 `isin()`이다. 이때는 리스트(list) 안의 모든 레코드를 필터링한다.

````python
# 임의 3 개의 entity 고유값을 가지는 행만 필터링
countries = ['South Korea', 'China', 'India']
subdf = df.loc[df['Entity'].isin(countries)]
subdf
````

> 자 이제 그림을 그리기 위한 라이브러리를 불러오자.

````python
import matplotlib.pyplot as plt
import seaborn as sns
````

> matplotlib에서 `subplots()`는 여러 그래프를 그릴 때 특히 유용하다. 비유하자면 한 페이지에 여러 컷이 들어있는 만화를 떠올리면 좋다.
>
> seaborn 라이브러리로 **라인 차트(line chart)**를 시각화할 때 `sns.lineplot()`을 사용한다. 시계열 도표는 본질적으로 라인 차트이기 때문이다. 이때 `hue` 패러미터를 사용하지 않으면 어떻게 되는지 실험해보자.

````python
#fig는 페이지, ax는 컷
fig, ax = plt.subplots()

#line chart
sns.lineplot(data=subdf,
             x='Year',
             y='No. of Internet Users',
             ax=ax,
             hue='Entity')                # Entity 변수별로 색조 구분
````

> 위 그래프에서 $y$축을 사용자 퍼센티지('Internet Users(%)')로 변경하여 다시 시각화해보자. `palette` 패러미터를 사용해보자. **범례(legends)**의 순서도 바꿀 수도 있다.

````python
#범례 순서
countries = ['South Korea', 'China', 'India']

#fig는 페이지, ax는 컷
fig, ax = plt.subplots()

#line chart
sns.lineplot(data=subdf,
             x='Year',
             y='Internet Users(%)',                 #이번엔 퍼센트
             ax=ax,
             hue='Entity', hue_order=countries,     #범례 순서
             palette='bright')                      #팔렛=bright, pastel, deep
````

> `hue` 대신 혹은 동시에 `style`을 사용해 국가를 구분할 수 있다. `markers` 패러미터를 사용하면 나름 강조에 편리하다.

````python
countries = ['South Korea', 'China', 'India']

#fig는 페이지, ax는 컷
fig, ax = plt.subplots()

#line chart
sns.lineplot(data=subdf,
             x='Year',
             y='Internet Users(%)',                 #이번엔 퍼센트
             ax=ax,
             hue='Entity', hue_order=countries,     #범례 순서
             palette='bright',                      #팔렛=bright, pastel, deep
             style='Entity',                        #모양으로 구분
             markers=['o','^','X'])
````

> 선의 두께를 조절할 수도 있다. 왜 중국을 나타내는 선이 두꺼워졌는지도 주의해서 살펴보자.

````python
countries = ['South Korea', 'China', 'India']

#fig는 페이지, ax는 컷
fig, ax = plt.subplots()

#line chart
sns.lineplot(data=subdf,
             x='Year',
             y='Internet Users(%)',                 #이번엔 퍼센트
             ax=ax,
             hue='Entity', hue_order=countries,     #범례 순서
             palette='bright',                      #팔렛=bright, pastel, deep
             #style='Entity',                       #모양으로 구분
             size='Entity', sizes=[3, 2, 1],        #두께
             #markers=['o','^','X']
             )
````

> **연습문제 2-1**. `tips` 자료에서 총 청구액(`total_bill`)과 팁(`tip`)의 연관성을 시각화하되, 저녁 시간인지 점심 시간인지에 따라 다르게 표현하시오. 아름답게 표현하기 위해 검색을 활용하시오.

````python
#tips data set
df = sns.load_dataset('tips')

#scatter plot
fig, ax = plt.subplots()
sns.scatterplot(data=df,
                x='total_bill',
                y='tip',
                ax=ax,
                hue='time',       #time에 따라 색조를 다르게
                style='time')     #time에 따라 스타일도 다르게
````

> 그런데 산점도를 그릴 때는 반드시 **적합선(fitted line)**을 함께 그려야 한다. 사실 적합선은 **회귀분석(regression analysis)**을 통해 얻어야 한다(Why?).

````python
#scatter plot
fig, ax = plt.subplots()
sns.regplot(data=df,
            x='total_bill',
            y='tip',
            ax=ax,
            ci=None)            # 신뢰구간 삭제
````

> 사실 세부적인 미세조정(fine-tuning)이 얼마든지 가능하다.

````python
#스타일 옵션 조정은 딕트로
s_styles = {'s' : 10}             #산점도의 점 크기
l_styles = {'color' : 'gray',     #적합선의 색깔과 선 스타일
            'linestyle' : '--'}

#scatter plot
fig, ax = plt.subplots()
sns.regplot(data=df,
            x='total_bill',
            y='tip',
            ax=ax,
            ci=None,          # 신뢰구간 삭제
            scatter_kws=s_styles,
            line_kws=l_styles)
````

> **연습문제 2-2**. 위 시각화를 좀 더 아름답게 표현하기 위해 검색을 활용하여 개선하시오.
---
https://seaborn.pydata.org/generated/seaborn.regplot.html

>
> 아까 배웠듯 `sns.scatterplot()` 안에서 `hue` 그리고/또는 `style` 패러미터를 사용해 색조/스타일을 달리 표현한 산점도를 먼저 그린다.
>
> 그 뒤, `sns.regplot()`을 덧씌워 그리고, `scatter=False` 패러미터를 준다!

````python
#scatter plot
fig, ax = plt.subplots()
sns.scatterplot(data=df,
                x='total_bill',
                y='tip',
                ax=ax,
                hue='time',       #색조 다르게
                style='time')     #스타일 다르게

#scatter plot
sns.regplot(data=df,
            x='total_bill',
            y='tip',
            ax=ax,
            ci=None,          # 신뢰구간 삭제
            scatter=False)
````

## 4. 상자-수염 도표

> **상자-수염(Box-Whiskers plot)**은 양적변수를 범주별로 구분지어 요약할 때 종종 사용된다. `EV_charge.csv`는 전기차 충전 이력 **로그 파일(log file)**이다.
>
> 상자-수염 도표에서는 종종 **이상치(outliers)** 때문에 고민하게 된다. `showfliers` 패러미터로 조절할 수 있다. 그런데 이때 이상치를 아예 삭제하고 시각화한다는 점에 주의해야 한다(Why?).

````python
df = pd.read_csv('EV_charge.csv')
df

fig, ax = plt.subplots()
sns.boxplot(data=df,
            x='weekday',
            y='kwhTotal',
            #showfliers=False,          #outliers
            ax=ax)
````

> 그런데 잘 살펴보면 $x$축이 제대로 정렬되어 있지 않다. 그러므로 `order` 패러미터를 설정해야 한다.

````python
weekday_order = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']

fig, ax = plt.subplots()
sns.boxplot(data=df,
            x='weekday',
            y='kwhTotal',
            ax=ax,
            order=weekday_order)
````

> Swarm plot은 상자-수염 그림의 특수한 변형이다. 자료가 크면 그리는데 오래 걸린다.

````python
weekday_order = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']

fig, ax = plt.subplots()

#strip box plot
sns.swarmplot(data=df,
              x='weekday',
              y='kwhTotal',
              ax=ax,
              #size=.5,       #사이즈를 줄이면 에러 메시지가 출력되지 않지만 이 그림을 원하지 않을수 있다.
              order=weekday_order)
````

> Seaborn stripplot과 boxplot 동시 활용

````python
weekday_order = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']

fig, ax = plt.subplots()

#box-whiskers plot
sns.boxplot(data=df,
            x='weekday',
            y='kwhTotal',
            ax=ax,
            showfliers=False,
            order=weekday_order)

#strip box plot
sns.swarmplot(data=df,
              x='weekday',
              y='kwhTotal',
              ax=ax,
              order=weekday_order,
              color='grey', alpha=0.4)
````

> 상자-수염 그림을 관심변수 유형별로 나란히 놓고 비교할 수도 있다. 이렇게 유형별로 묶는 방식을 **그룹별 상자-수염 도표(Grouped box-whiskers plot)**라고 부른다.
>
> 가령 플랫폼(`platform`)별로 그룹지어 시각화해보자. 우선 플랫폼 빈도분포표를 확인해보자. web은 너무 빈도가 적으므로 아예 삭제하는 것이 어떨까?

````python
df['platform'].value_counts()
noweb = df.loc[df['platform'] != 'web']
````

````python
weekday_order = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']

fig, ax = plt.subplots()

#box-whiskers plot
sns.boxplot(data=noweb,
            x='weekday',
            y='kwhTotal',
            ax=ax,
            showfliers=False,          #outliers
            order=weekday_order,
            hue='platform')            #플랫폼별로 색조 다르게
````

> **연습문제 4**. 위 자료에서 적절한 변수를 골라 상자-수염 도표를 시각화하시오. 적절한 검색을 통해 아름답게 꾸미시오.

````python
#no web
noweb = df.loc[df['platform'] != 'web']
noweb['facilityType'].value_counts()

fig, ax = plt.subplots()

#box-whiskers plot
sns.boxplot(data=noweb,
            x='platform',
            y='chargeTimeHrs',
            ax=ax,
            showfliers=False,          #outliers
            hue='facilityType')            #플랫폼별로 색조 다르게
````

## 5. 인터렉티브 시계열 도표

> plotly는 **인터렉티브 그래프(interactive graph)**를 만들 때 현업에서도 활용될 정도로 아주 편리하고 강력한 라이브러리이다. 사용법도 간단하다.
>
> seaborn과는 다소 사용법이 다르기 때문에 다시 시계열 도표를 만들어보자.

````python
import plotly.express as px
import pandas as pd

#아시아 3개국
df = pd.read_csv('global_internet_users.csv')
countries = ['South Korea', 'China', 'India']
subdf = df.loc[df['Entity'].isin(countries)]

#시계열 도표
fig = px.line(data_frame=subdf,
              x='Year',
              y='Internet Users(%)',
              width=600, height=400,
              color='Entity')         #Entity별로 색조를 다르게
fig.show()
````

> 선 모양을 바꿀 때는 `line_dash` 패러미터를 사용한다.

````python
fig = px.line(data_frame=subdf,
              x='Year',
              y='Internet Users(%)',
              width=600, height=400,
              line_dash='Entity',         #Entity로 선을 다르게
              color='Entity')             #Entity별로 색조도 다르게
fig.show()
````

> 각각의 관측치에 따라 심볼(symbol)을 다르게 할 수도 있다.

````python
fig = px.line(data_frame=subdf,
              x='Year',
              y='Internet Users(%)',
              width=600, height=400,
              line_dash='Entity',         #Entity별로 선을 다르게
              color='Entity',             #Entity별로 색을 다르게
              symbol='Entity')            #Entity별로 심볼을 다르게
fig.show()
````

## 6. 인터렉티브 산점도

> 산점도 역시 seaborn 뿐 아니라 plotly에서 구현할 수 있다. 앞서 사용했던 `tips` 자료를 다시 한 번 사용하여 이를 구현해보자.

````python
#tips data set
df = sns.load_dataset('tips')

fig = px.scatter(data_frame=df,
                 x='total_bill',
                 y='tip',
                 width=600, height=400)
fig.show()
````

> `color` 패러미터를 사용하면 관심있는 변수(가령 `time`)에 따라 다른 색조로 산점도를 그릴 수 있다.

````python
fig = px.scatter(data_frame=df,
                 x='total_bill',
                 y='tip',
                 width=600, height=400,
                 color='time')            #time에 따라 색조를 다르게
fig.show()
````

> 필요하다면 **컬러 스킴(color scheme)**이 따로 있으므로 가져다 쓸 수 있다.
---
https://plotly.com/python/discrete-color/

````python
#color scheme
color = px.colors.qualitative.D3

fig = px.scatter(data_frame=df,
                 x='total_bill',
                 y='tip',
                 width=600, height=400,
                 color='time', color_discrete_sequence=color)
fig.show()
````

## 7. 인터렉티브 히스토그램

> 히스토그램도 plotly로 구현할 수 있다. 이 경우에는 `histogram()`을 사용한다.

````python
df = sns.load_dataset('tips')

fig = px.histogram(data_frame=df,
                   x='total_bill',
                   nbins=20,
                   width=600, height=400)
fig.show()
````

> 막대의 색을 바꾸거나 막대 간 간격을 적절히 떼어놓을 수 있다. 구체적인 것은 검색과 ChatGPT의 도움이 필요하다.

````python
fig = px.histogram(data_frame=df,
                   x='total_bill',
                   nbins=20,
                   width=600, height=400,
                   color_discrete_sequence=['deeppink'])      #막대 색깔
fig.update_layout(bargap=0.2)                                 #막대간 간격
fig.show()
````

> 그룹별로 히스토그램을 나누어 그릴 수도 있다. 여러 히스토그램을 하나의 그래프 안에 합치는 방법이 몇 가지 있는데, `barmode` 패러미터 안에서 바꿀 수 있다. 이때부터 컬러 스킴을 설정할 필요가 생겨난다.

````python
#color scheme
color = px.colors.qualitative.D3

fig = px.histogram(data_frame=df,
                   x='total_bill',
                   nbins=20,
                   width=600, height=400,
                   color='time', barmode='relative',          #'relative', 'overlay' 또는 'group'
                   color_discrete_sequence=color)             #막대 색깔
fig.update_layout(bargap=0.05)                                #막대간 간격
fig.show()
````

## 8. 인터렉티브 상자-수염 도표

> plotly로 그리는 상자-수염 도표는 `px.box`로 그린다.
>
> 이때에도 이상치(outliers) 문제가 다시 한 번 부상할 수 밖에 없다. `points` 패러미터로 조정가능하지만 조금 복잡하다. 이상치 표현은 네 가지로 설정이 가능하며 하나하나 연습해야 한다.

````python
df = pd.read_csv('EV_charge.csv')
noweb = df.loc[df['platform'] != 'web']

# Plotly boxplot
fig = px.box(data_frame=noweb,
             x='weekday',
             y='kwhTotal',
             points='outliers',            #이상치('outliers', 'suspectedoutliers', 'all', or False)
             width=600, height=400)
fig.show()
````

> 아까와 마찬가지로 $x$축 순서가 틀렸다. 그런데 그 수정 방법이 약간 특이하다.

````python
weekday_order = {'weekday': ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']}   #딕트

# Plotly boxplot
fig = px.box(data_frame=noweb,
             x='weekday',
             y='kwhTotal',
             points='suspectedoutliers',      #이상치('outliers', 'suspectedoutliers', 'all', or False)
             category_orders=weekday_order,
             width=600, height=400)
fig.show()
````

> 마지막으로 그룹별 상자-수염 도표를 그려야 한다. 이때도 크게 다르지 않다. `color` 패러미터로 간단히 그룹을 설정할 수 있기 때문이다.

````python
weekday_order = {'weekday': ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']}   #딕트

# Plotly boxplot
fig = px.box(data_frame=noweb,
             x='weekday',
             y='kwhTotal',
             #points='outliers',                #이상치('outliers', 'suspectedoutliers', 'all', or False)
             category_orders=weekday_order,
             color='platform',                  #플랫폼 별로 색조를 다르게
             width=600, height=400)
fig.show()
````

## 9. 컷의 조절

> 하나의 페이지에 여러 컷의 그림을 넣어보자.
>
> 먼저 `medical_cost.csv` 자료를 불러온 뒤 내용을 살펴보자. 지역(`region`)의 빈도분포표를 검토하자.

````python
# matplotlib 2x2 subplot에 Seaborn regplot 시각화
df = pd.read_csv('medical_cost.csv')
df
df['region'].value_counts()
````

> matplotlib의 `subplots()`을 단순히 `fig`, `ax` 에 집어넣지 말고, 좀 더 구체적으로 지정한다. 2x2 그림을 그리고 페이지 크기도 설정한다.

````python
fig, ax = plt.subplots(2, 2, figsize=(12, 12))
````

> 우리는 총 4개의 컷을 그려넣어야 하므로(Why?), `ax[0][0]`에서 `ax[1][1]`까지 차례로 집어넣어 그린다.
>
> 또한 각 `ax[i][j].set_title()` 매서드로 컷의 제목도 넣자!

````python
#2x2 그림, 페이지 크기도 설정
fig, ax = plt.subplots(2, 2, figsize=(12, 12))

#좌상단
sns.regplot(x='bmi',
            y='charges',
            data=df.loc[df['region'] == "northwest"],
            ax=ax[1][0])                    #어디에 그려넣을지 여기서 설정
ax[0][0].set_title('Northwest')

#우상단
sns.regplot(x='bmi',
            y='charges',
            data=df.loc[df['region'] == "northeast"],
            ax=ax[1][1])
ax[0][1].set_title('Northeast')

#좌하단
sns.regplot(x='bmi',
            y='charges',
            data=df.loc[df['region'] == "southwest"],
            ax=ax[0][0])
ax[1][0].set_title('Southwest')

#우하단
sns.regplot(x='bmi',
            y='charges',
            data=df.loc[df['region'] == "southeast"],
            ax=ax[0][1])
ax[1][1].set_title('Southeast')
````

> 이렇게 그리면 매 **축(axis)**마다 그림을 그려야 한다는 번거로움은 있다. 그리고 $x$축과 $y$축이 그림마다 다르다는 단점도 있다(나중에 배우겠지만 사실 이 문제들은 그다지 심각한 문제가 아니다).
>
> 한편 어떤 사람들은 명령어 하나로 간단히 그리길 원한다. 만약 산점도와 적합선의 경우라면 seaborn의 `lmplot()`이 사용될 수 있다.

````python
sns.lmplot(data=df,
           x='bmi',
           y='charges',
           col='region',      #region별로 컬럼을 나눔
           col_wrap=2)        #컬럼은 2개씩 그리고 한줄 내림
````

> `hue` 패러미터를 사용하여 자료의 성격(가령 `smoker`)에 따라 다른 두 그림을 하나의 그래프 안에 넣을 수 있다. 이것은 자료의 이질성을 시각화할 때 당연히 중요하게 활용된다!

````python
sns.lmplot(data=df,
           x='bmi',
           y='charges',
           hue='smoker')
````

> **연습문제 9-1**. $x$축을 `bmi`, $y$축을 `charges`로 하는 산점도와 적합선을 그리되, 각각의 성별(`sex`)로 나누어 두 개의 그래프를 상하로 그리시오.

````python
sns.lmplot(data=df,
           x='bmi',
           y='charges',
           col='sex', col_wrap=1)
````

> plotly를 사용하여 산점도와 적합선을 그릴 때는 `scatter()`를 사용한다. 이때는 `facet_col` 패러미터를 이용하여 그룹별로 **서브플롯(subplot)**에 나누어 그릴 수 있다.


````python
fig = px.scatter(data_frame=df,
                 x='bmi',
                 y='charges',
                 facet_col='region', facet_col_wrap=2,       #region별로 컬럼을 나누기
                 width=800, height=500,
                 trendline='ols')             #적합선
fig.show()
````
