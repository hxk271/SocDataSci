## 2. 마크다운과 텍스트

> 이제 퍼블리시 방법을 익혔으니 계속해서 Streamlit 코드를 만들어보자. 이번엔 **마크다운(markdown)**을 추가해보자. 웹페이지를 만들 때는 마크다운 문법을 사용하게 된다. 마크다운의 문법은 여러 책과 참고자료가 인터넷에 많으므로 각자가 알아서 공부하자. 특히 `마크다운 핵심 요약` 따위의 검색어가 유용하다. 마크다운이나 일반적인 텍스트 이외에 **코드 블럭(code block)**을 넣을 수도 있다. 별건 아니지만 **디바이더(divider)**가 의외로 구획을 깨끗하게 나눠주기 때문에 유용하다.

````python
%%writefile app.py

import streamlit as st

st.title('This is title')
st.header('This is header')
st.subheader('This is subheader')

# 마크다운
doc = '''
    메인 텍스트
    색깔을 바꾸고 싶다구요? :red[빨간색], :blue[파란색], :green[녹색]입니다.
    **볼드**나 *이탤릭*도 됩니다.
    '''
st.markdown(doc)

# 마크다운 아닌 text
st.text(doc)

# -----------------------------
st.divider()

# 코드 블럭
code = '''
[여기에 여러분의 파이썬 코드를 복사해서 붙여넣어보자!]
'''
st.code(code, language='python')
````

## 3. 다양한 입력 위젯

> 인터렉티브 대시보드에는 **입력 위젯(input widget)**이 꼭 필요하다(Why?). 모든 입력 위젯의 첫번째는 누가 뭐래도 **버튼(button)**이다. 딱봐도 누르고 싶게 디자인해야 한다. `st.button()`으로 만든다.

````python
%%writefile app.py

import streamlit as st

# button
def button_write():
    st.markdown('쾅~')

#버튼 1
st.button('눌르지 마시오',           #안에 나오는 글자
          type='primary',            #빨갛게 주목
          on_click=button_write)     #눌리면 button_write 함수 작동

#버튼 2
st.button('누르시오',
          type='secondary')          #사실 이게 디폴트

#버튼 3
if st.button('Reset')==True:         #누르면 이 조건이 성립!
    st.text('오케')
````

> 이제 **체크박스(check box)** 그리고 그와 비슷한 **토글(toggle)**을 만들어보자. 체크박스는 (체크되어 있지 않은) 비활성화가 디폴트임에 주의하자. `markdown`과는 달리 `text`는 엉뚱한 곳에 나타남을 확인하자. `if`를 사용하는 방식과 함수를 사용하는 방식 두 가지를 모두 기억하자.

````python
%%writefile app.py

import streamlit as st

st.title('출생 신GO')
st.header('국민의례')

# 마크다운
st.markdown("나는 자랑스러운 태극기 앞에 ~~~~~~~~~ 다짐합니다.")
st.divider()

# checkbox 1
active = st.checkbox('I Agree!')
if active == True:
    st.markdown('웰컴 투 **코리아**')
st.divider()

# checkbox 2
def checkbox_write():
    st.text('웰컴~ 웰컴~')
st.checkbox('I agree', on_change=checkbox_write)
st.divider()

# toggle
toggle = st.toggle('당신은 의무를 다하시겠습니까?',
                   value=True)          #value=True는 활성화가 디폴트
if toggle == True:
    st.text('그럼 좋습니다')
else:
    st.text('헐 그럼 안되지~')
````

> 이제 좀 더 다양한 옵션 가운데 하나 이상을 골라보자. **셀렉트 박스(selectbox)**, **라디오(ratio)**, **멀티셀렉트(multiselect)**가 유용하다.

````python
%%writefile app.py

import streamlit as st

# selectbox
option = st.selectbox(label='앞으로 커서 탈 차를 골라보시오',
                      options=['현차', '기차', 'BMW', '벤츠', '아우디', '그 외'],
                      index=None,              # None이 디폴트
                      placeholder='없음',      # 디폴트로 보여줄 선택지
                      disabled = False)        # True로 바꾸면 비활성화되어 고를 수 없다
st.text(str(option) + '를 탄다구요?')
st.divider()

# radio
option = st.radio(label = '앞으로 좋아할 영화 장르는 무엇인가요?',
                  options = ["코미디", "드라마", "다큐"],
                  captions = ['웃으며 삽시다', '너무 울지마세요', '랜선 집사'])

if option:                                  # option이 뭐라도 골라졌으면
    st.markdown(option + '를 골랐군요.')
st.divider()

# multiselect
option = st.multiselect(label = '앗차, 여러개 골라도 됩니다!',
                        options = ["코미디", "드라마", "다큐"],
                        placeholder='좋아하는 것은?')
st.markdown(str(option) + '를 골랐군요.')           # none 때문에 str(option)
````

> **연습문제 3-2**. 위의 코드를 수정하고자 한다. 먼저 의무를 다하겠다고 고른 경우 앞으로 탈 차를 고르도록 허용하시오. 그리고 앞으로 탈 차가 '그 외'인 경우, 이번엔 주된 운송수단까지 고를수 있도록 설계하시오. 이때 주된 운송 수단으로 고를 수 있는 옵션은 Bus, Metro, Walk으로 설정하고, 디폴트로 출력할 옵션은 Walk로 나오도록 만드시오.

````python
%%writefile app.py

import streamlit as st

st.title('출생신GO')
st.header('국민의례')

# 마크다운
doc = '''
    나는 자랑스러운 태극기 앞에 ~~~~~~~~~ 다짐합니다.
    '''
st.markdown(doc)

# divider
st.divider()

# checkbox 1
active = st.checkbox('I Agree!')
if active == True:
    st.markdown('웰컴 투 **코리아**')

# toggle
toggle = st.toggle('의무를 다하시겠습니까?',
                   value=False)          #value=True는 켜짐 상태가 디폴트
st.divider()
if toggle == True:

    # selectbox
    option = st.selectbox(label='앞으로 커서 탈 차를 골라보시오',
                        options=['현차', '기차', 'BMW', '벤츠', '아우디', '기타'],
                        disabled = False)        # True로 바꾸면 비활성화되어 고를 수 없다
    st.divider()

    # 기타를 고른 경우
    if option == '기타':

        # selectbox placeholder
        option = st.selectbox(label='주된 운송수단을 골라보시오',
                            options=['Bus', 'Metro', 'Walk'],
                            index=2,                   # 2가 Walk
                            placeholder='없음')        # 디폴트로 보여줄 선택지
        st.markdown(str(option) + '를 골랐네요.')   # 왜 str(option) 일까?
````

> 날짜와 시간을 슬라이더로 입력받을 수 있다. 그에 앞서 datetime 라이브러리를 먼저 공부해보자.

````python
from datetime import time

# 시간
timer = time(23, 0, 57)
print(timer)
print(timer.hour)
print(timer.minute)
print(timer.second)

# length of interval (LOIs)
t1 = time(14, 12, 1)
t2 = time(14, 14, 0)
#print(t1-t2)                 # TypeError: unsupported operand type(s) for -: 'datetime.time' and 'datetime.time'

# 제대로 하기
from datetime import timedelta

delta = timedelta(hours=t2.hour-t1.hour,
                  minutes=t2.minute-t1.minute,
                  seconds=t2.second-t1.second)
print(delta)
````

````python
%%writefile app.py

import streamlit as st
from datetime import time

t1, t2 = st.slider(label = '기말시험이 머지 않았습니다. 평균적으로 언제부터 언제까지 공부하겠습니까?',
                   min_value=time(0),
                   max_value=time(23),
                   value=(time(8), time(18)),      # 디폴트로 줄 시간대. set 형식
                   format='HH:mm')                 #포멧!

#공부 시간 LOIs 계산
st.markdown(str(t2.hour - t1.hour) + '시간 공부하는군요!')
````

> **연습문제 4-3**. 12주차 **연습문제 1-1**을 대시보드로 출력해보자. 이때 이용자가 직접 분석할 수 있도록 최대한의 인터렉티브한 환경을 제공하자. seaborn 그림과 Plotly 그림을 각각 연습해보자.

````python
%%writefile app.py

import streamlit as st
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('C:/Users/user/Downloads/ds_salaries.csv')

# groupby() 입력받기
grpby = st.selectbox(label='Select variable for groupby',
                     index=3,     #none이면 아래 groupby()에서 none이 들어가므로 오류가 난다!
                     options=df.columns)

# 평균값을 확인할 변수 입력받기
xvar = st.selectbox(label='Select statistics to report',
                    index=6,            #none이면 아래에서 none이 들어가므로 오류가 난다!
                    options=df.columns)

# 상위 몇 명을 확인할지 입력받기
count = st.slider(label = 'Select how many to report',
                  min_value = 1, max_value = 20,
                  value = 10)          #연습문제 1-1에서 디폴트는 10

job_salary = df.groupby(grpby)[xvar].mean()
tops = job_salary.sort_values(ascending = False)[0:count]
tops = tops.reset_index()

fig, ax= plt.subplots()
sns.barplot(data=tops,
            y=grpby,
            x=xvar,
            ax=ax)

# 일반적인 plt 그림을 대시보드에 출력
st.pyplot(fig)
````
